````python
from yahoo_api import YahooApi
from etoro_api import EtoroApi
import pandas as pd
from yahoo_finance_api2 import share
yahoo_api = YahooApi()
etoro_api = EtoroApi()
pd.set_option('display.max_rows', 500)
````

````python
selected_stock = ""
for stock in etoro_api.get_stocks(get_all=True)["SymbolFull"]:
    if (stock == "TQQQ"):
        print(stock)
        selected_stock = stock
````

Class Share:
get_historical(period_type, period, frequency_type, frequency)
Returns historical data for the given 
period type 
```
(share.PERIOD_TYPE_DAY, share.PERIOD_TYPE_WEEK, share.PERIOD_TYPE_MONTH, share.PERIOD_TYPE_YEAR),
```
```
period (1, 5, 10, etc.), 
```
```
frequency_type (share.FREQUENCY_TYPE_MINUTE, share.FREQUENCY_TYPE_DAY, share.FREQUENCY_TYPE_MONTH, share.FREQUENCY_TYPE_YEAR)
```
```
frequency (1, 5, 10, etc.).
``` 
Only certain combinations of these parameters are allowed.

````python
df = yahoo_api.get_stock_data(selected_stock, periods=50, take_profit=1.05, intervals_to_profit=2, period_type=share.PERIOD_TYPE_YEAR, frequency_type=share.FREQUENCY_TYPE_DAY, frequency=1)
````

````python
# df_copy = df.copy()
# df["profit"] = (df_copy.loc[df["rowIndex"]:df["rowIndex"]]["Close"] > 100).any()
def check_for_profit(df, minutes = 30):
    for i in range(minutes):
        if df.shift(minutes * -1)["Close"] > 100:
            return "True"
    return "False"
````

````python
# Returns true, when `Close` goes up 1.5 % in the next 30 minutes
df["profit"] = df.apply(lambda x: (x["volume"] > 0) & ((df.loc[x["rowIndex"]:(x["rowIndex"]+30)]["Close"] > x["Close"] * 1.015).any()) if True else False, axis=1)
````
